### Outlier Detection: A quick note

1. One general rule that could be followed is to assume: 

$$ Outlier = 3 * z$$

where

$$ z = \frac {x_i -\mu} \sigma $$

$$x = observation $$
$$\mu = mean $$
$$\sigma = stddev$$

2. The seconds approach could be using the IQR The interquartile range (IQR), also called the midspread or middle 50%, or technically H-spread, is a measure of statistical dispersion, being equal to the difference between 75th and 25th percentiles, or between upper and lower quartiles, IQR = Q3 − Q1. It is a measure of the dispersion similar to standard deviation or variance, but is much more robust against outliers. IQR is somewhat similar to Z-score in terms of finding the distribution of data and then keeping some threshold to identify the outlier.

```python
detailed_explanation = 'https://mathbitsnotebook.com/Algebra2/Statistics/normalstandard.jpg'
iframe = '<iframe src={} width=800 height=500></iframe>'.format(detailed_explanation)
HTML(iframe)
```

#### Example 1: Outlier removal using IQR Range

```python
Q1 = num_df.quantile(0.25)
Q3 = num_df.quantile(0.75)
IQR = Q3 - Q1
print('Here is our 25th percentile assoc with each field: \n')
print(Q1)
print('\n')
print('Here is our 75th percentile assoc with each field: \n')
print(Q3)
print('\n')
print('Here is our interquartile range assoc with each field: \n')
print(IQR)
print('\n')
print(type(IQR))
```

```python
num_df.shape
```

What we do next is apply a simple filtering condition to say modify our dataframe by filtering out the values `pipe` operator `|` means `OR`

```python
num_df_out = num_df[~((num_df < (Q1 - 1.5 * IQR)) | (num_df > (Q3 + 1.5 * IQR))).any(axis=1)]
num_df_out.shape
```

**Observations**

1. As you can see we've lost quite a number of records in our DataFrame. 
2. This is not typicaly something we do as outliers often do give us interesting insights. When in doubt, keep your data.

### Data Imputation

**Data Imputation is not advised, especially if you're imputing over 10% of a features values**

The general rule I follow is to drop records unless I can find an apropriate model for imputing data. For example if a have a large chunk of missing data associated with a feature, and I can find another feature/features that; for example is linearly correlated with that feature, to develop a regression model that fits new points to missing values in my sparse feature. Another option is linear interpolation for time series data and so forth
